# Vegetation management strategies

Vegetation growth may cause water levels to increase above certain thresholds. To prevent this, vegetation may be removed. This can be simulated through the event configuration file. 

```python
from dotter.models import DotterModel
%matplotlib notebook

no_events = DotterModel('config/vegetation_config.ini', loggerlevel='silent')
no_events.run()
no_events.dash(2);
```

```python
with_events = DotterModel('config/events_config.ini', loggerlevel='silent')
with_events.run()
with_events.dash(2);
```
